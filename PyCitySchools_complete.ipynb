## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
reduced_school_df = school_df.loc[:, ["school_name", "reading_score", "math_score", "% Passing Math", 
                                      "% Passing Reading", "Overall Passing Rate"]]

bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

reduced_school_df["$/Student Groups"] = pd.cut(school_df["$/student"], bins, labels=group_names)
reduced_school_df.head()


school_spending_group= reduced_school_df.rename(columns={"school_name": "School Name", "reading_score": "Reading Score", "math_score": "Math Score"})
#school_spending_group.round(2)
```

```python
spend_group = school_spending_group.groupby("$/Student Groups")
spend_group[['Reading Score', 'Math Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']].mean().round(2)
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_school_df = school_df.loc[:, ["school_name", "reading_score", "math_score", "% Passing Math", 
                                      "% Passing Reading", "Overall Passing Rate"]]

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

size_school_df["School Size Groups"] = pd.cut(school_df["size"], size_bins, labels=group_names)
size_school_df.head()


size_groups = size_school_df.rename(columns={"school_name": "School Name", "reading_score": "Reading Score", "math_score": "Math Score"})
#size_groups.round()
```

```python
size_group = size_groups.groupby("School Size Groups")
size_group[['Reading Score', 'Math Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']].mean().round(2)
```
